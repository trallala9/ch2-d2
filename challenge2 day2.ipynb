```python
from keras.applications.resnet50 import ResNet50, decode_predictions, preprocess_input
from keras.preprocessing import image
import numpy as np

import requests 
from io import BytesIO
from PIL import Image
```

```python
model = ResNet50(weights='imagenet')
```

![](https://calicoxtabby.files.wordpress.com/2015/04/images-2.jpg)

```python
url_img = 'https://www.sciencenews.org/sites/default/files/2019/02/main/articles/022119_EC_expanding-universe_feat.jpg'

response = requests.get(url_img)
img = Image.open( BytesIO(response.content) )
img = img.resize((224, 224))
img
```

```python
X = image.img_to_array(img)
X = np.expand_dims(X, axis = 0)
X.shape
```

```python
y_pred = model.predict(X)
```

```python
decode_predictions(y_pred, top=3)
```
